```python
"""Corpus	A collection of text used for NLP	example:Gutenberg corpus in NLTK
Stopwords	Common words removed from text	example:"is", "the", "in"
Stemming	Reducing words to their root form example:	"running" → "run", "flies" → "fli"
PorterStemmer	A popular stemming algorithm  example	"playing" → "play", "happiness" → "happi"
"""
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
import string 
```

```python
"""

transform_text function flow 

Converts to Lowercase:

Ensures uniformity by converting the entire text to lowercase.
Example: "Playing is Fun" → "playing is fun"
Tokenization (Splitting Words):

Uses NLTK’s word_tokenize to split text into individual words/tokens.
Example: "Hello! How are you?" → ['Hello', '!', 'How', 'are', 'you', '?']
Removes Non-Alphanumeric Characters:

Removes punctuation and special symbols, keeping only words and numbers.
Example: ['Hello', '!', 'How', 'are', 'you', '?'] → ['Hello', 'How', 'are', 'you']
Removes Stopwords & Punctuation:

Filters out common words like "is", "the", "and" (stopwords) and punctuation.
Example: ['Hello', 'How', 'are', 'you'] → ['Hello', 'How']
Applies Stemming Using PorterStemmer:

Converts words to their root form (removes suffixes like -ing, -ed, -es).
Example: ['playing', 'running', 'easily'] → ['play', 'run', 'easili']
Stores Intermediate Results in text = y[:]:

Creates a copy of y to prevent unwanted modifications.
y.clear() ensures y is empty before using it in the next step.
Joins Processed Words into Final Output String:

Uses " ".join(y) to return cleaned and transformed text.
Example Output: "play run easili"
"""
```

```python
def transform_text(text):
    text = text.lower()
    text = nltk.word_tokenize(text)
    
    y = []
    for i in text:
        if i.isalnum():
            y.append(i)
    
    text = y[:]
    y.clear()
    
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
            
    text = y[:]
    y.clear()
    
    for i in text:
        y.append(ps.stem(i))
    
            
    return " ".join(y)
```

```python
df['transformed_text']=df['text'].apply(transform_text)
```

```python
"""df[df['target'] == 0] → Filters only ham messages (target == 0 means non-spam).
['transformed_text'] → Selects the column containing preprocessed (cleaned) text. 
 
 both line will give a series of line 
.tolist() → Converts the column into a Python list of messages.This will put all lines in a single list 
Outer Loop: Iterates through each ham message (msg).
Inner Loop: Splits each message line into words (msg.split())  and adds them to ham_corpus
"""

ham_corpus = []
for msg in df[df['target'] == 0]['transformed_text'].tolist():
    for word in msg.split():
        ham_corpus.append(word)
        
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
cv=CountVectorizer()
tfidf=TfidfVectorizer(max_features=3000)
```

```python
X=tfidf.fit_transform(df['transformed_text']).toarray()
```

```python
X.shape
```

```python
y=df['target'].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score

# Dictionary of models
models = {
    "GaussianNB": GaussianNB(),
    "MultinomialNB": MultinomialNB(),
    "BernoulliNB": BernoulliNB(),
    "LogisticRegression": LogisticRegression(),
    "DecisionTree": DecisionTreeClassifier()
}
```

```python
# Loop through each model
for name, model in models.items():
    model.fit(X_train, y_train)  # Train model
    y_pred = model.predict(X_test)  # Predict
    # Print results
    print(f"\nModel: {name}")
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print(f"Precision: {precision_score(y_test, y_pred):.4f}")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
```
